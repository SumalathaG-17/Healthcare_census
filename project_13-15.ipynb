### Multi-line header - problem 13

```python
gHosp_df = pd.read_csv('../government_hospitals.csv',  header=[0,1])
gHosp_df.head()
```

```python
gHosp_df.columns
```

```python
gHosp_df.columns = ['State/UT', 'Rural_Government_Hospitals', 'Rural_Government_Beds', 
                    'Urban_Government_Hospitals', 'Urban_Government_Beds', 'Last_Updated']
gHosp_df.head()
```

### Data update and code reuse - problem 14

```python
gHosp_df['Last_Updated'] = pd.to_datetime(gHosp_df['Last_Updated'], format='%d.%m.%Y')
gHosp_df['Last_Updated'] = gHosp_df['Last_Updated'].dt.strftime('%Y-%m-%d')#fromatstringtime
gHosp_df.head()
```

```python
def format_states(df):
    name_correction = {"puducherry": "pondicherry", "nct of delhi": "delhi", "odisha": "orissa"}
    
    df["State/UT"]=df["State/UT"].str.lower()
    df["State/UT"]=df["State/UT"].replace(name_correction)
    df["State/UT"]=df["State/UT"].str.title()
    df['State/UT'] = df['State/UT'].str.replace('&', 'and').str.rstrip('*')
    return df
```

```python
gHosp_df = format_states(gHosp_df)
gHosp_df.head()
```

```python
gHosp_df.to_csv('../cleandata/government_hospital.csv', index=False)
```

### Government healthcare facility disparity - problem 15

*From the previous analysis it has been found that the states *['Bihar', 'Jharkhand', 'Uttar Pradesh']* has the least number of hospitals(all) wrt its population*

```python
gHosp_df['Total_gHospitals'] = gHosp_df['Rural_Government_Hospitals'] + gHosp_df['Urban_Government_Hospitals']
gHosp_df.head()
```

```python
st_with_least_hosp = ['Bihar', 'Jharkhand', 'Uttar Pradesh']
st_with_least_gHosp_df = gHosp_df[gHosp_df['State/UT'].isin(st_with_least_hosp)].sort_values(by='Total_gHospitals').head(1)
st_with_least_gHosp = st_with_least_gHosp_df.iloc[0]['State/UT']
print(f'The state with the least number of govt hospitals among {st_with_least_hosp} is {st_with_least_gHosp}.')
```
